```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first').fit(X_cat)

cols = encoder.get_feature_names_out(input_features=X_cat.columns)

X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)

X_cat_encode.head()
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(X, y, test_size=.2)
```

```python
#import library
from sklearn.linear_model import LinearRegression as linreg

lm = linreg()    # model
lm.fit(X_train, y_train)   # model training
y_pred = lm.predict(X_test)   # model prediction
```

```python
print ('train R2: {} -- test R2: {}'.format(lm.score(X_train, y_train),
                                            lm.score(X_test, y_test)))
```

```python
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

# Hypothesis

```python
#H0: There is no difference in the average temperature change across continents for the year 2022.
#H1: There is a difference in the average temperature change across continents for the year 2022.
```

```python
from scipy import stats


# Define the specific value for the t-test
stats_value = 0

for continent in df['continent'].unique():
    continent_data = df[df['continent'] == continent]['2022']
    
    # Perform one-sample t-test
    stat, pvalue = stats.ttest_1samp(continent_data, stats_value)
    
    # Display the result for each continent
    print(f"T-test for {continent} against {stats_value}:")
    print("Stat_Score =", stat)
    print("Pvalue =", pvalue)
    
    # Interpret the result
    if pvalue < 0.05:
        print("Reject the null hypothesis - there is a  difference.")
    else:
        print("Fail to reject the null hypothesis - no difference observed.")
    

   
```
