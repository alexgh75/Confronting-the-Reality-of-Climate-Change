```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Annual_Surface_Temperature_Change.csv")
```

```python
df
```

```python
cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower().replace(' ', '_'))
df.columns = cols
```

```python
# Remplacer les valeurs NaN par 'unknown' dans tout le DataFrame
df.fillna('unknown', inplace=True)

df.head()
```

```python
nan_values = df.isna().sum()
print(nan_values)
```

```python
categorical = df.select_dtypes(include=object)
numerical = df.select_dtypes(include=np.number)
```

```python
df.info()
```

```python
# Convert columns to numeric, excluding non-convertible values (e.g., 'unknown')
numerical_columns = df.columns[df.dtypes == 'object'][1:]  # Exclude the first column (assuming it's 'objectid')
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Display the updated DataFrame
print(df.dtypes)
```

```python
# Enlever la lettre 'f' des noms de colonnes des années
df.columns = [int(col[1:]) if col.startswith('f') else col for col in df.columns]

# Afficher le DataFrame après la modification
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate mean temperature change for each meteorological year
mean_temperature_change = df.loc[:, 'F1961':'F2022'].mean()

# Create a bar plot (histogram)
plt.figure(figsize=(12, 20))
sns.barplot(x=mean_temperature_change.index, y=mean_temperature_change.values, palette='viridis')
plt.title('Mean Temperature Change Over Meteorological Years')
plt.xlabel('Meteorological Year')
plt.ylabel('Mean Temperature Change (°C)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.show()

```
